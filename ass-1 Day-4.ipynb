```python
# print the first armstrong number in the range of 1042000 to 702648265 and exit the loop as soon as u encounter 
#the first armstrong number (Using while loop)
```

```python
for num in range (1042000, 702648266):
    temp = num
    sum = 0
    while temp > 0:
        digit = temp % 10
        sum = sum + (digit ** 3)
        temp = temp//10
        if temp == sum:
            print(num)
            break
```
